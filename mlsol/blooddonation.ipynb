```python
print(df_train.columns)
df_train.drop(['Unnamed: 0'],axis=1)
```

```python
x_train = df_train.iloc[:, df_train.columns != 'Made Donation in March 2007']
y_train = df_train.iloc[:, df_train.columns == 'Made Donation in March 2007']
```

```python
print("x: \n{} \ny_train : \n{}".format(x_train.head(3), y_train.head(3)))
```

```python
# Normalizing the Total Vloumne Donated Column because its values are very high as compared 
# to other column and will add biasness to this column
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
x_train['norm_Total Volume Donated (c.c.)'] = scaler.fit_transform(x_train['Total Volume Donated (c.c.)'].values.reshape(-1,1))
```

```python
from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test = train_test_split(x_train,y_train, test_size=0.2, random_state=584) 
```
